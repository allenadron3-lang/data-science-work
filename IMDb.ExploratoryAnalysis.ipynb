# IMDb Data Analysis

Simple example to read JSON data of the IMDb and calculate average ratings for a given actor.

```python
import json
import pandas as pd
```

```python
actors_id_name_map = {}
actors_count = {}
genres_count = {}
film_ratings = []
film_votes = []

# open the ratings data
with open("/Users/adronallen/Desktop/INST414/imdb_movies_2000to2022.prolific.json", "r") as in_file:
    
    # each line in the file is a JSON item
    for line in in_file:
        
        # read this movie
        movie = json.loads(line)
        
        # Iterate over each actor
        for actor_id,actor_name in movie["actors"]:
            actors_id_name_map[actor_id] = actor_name
            actors_count[actor_id] = actors_count.get(actor_id, 0) + 1

        # Iterate over each genre
        for genre in movie["genres"]:
            genres_count[genre] = genres_count.get(genre, 0) + 1

        if len(movie["rating"]) > 0:
            # Add the movie rating to the list
            film_ratings.append((movie["title"], movie["rating"]["avg"]))
    
            # Add the movie rating COUNT to the list
            film_votes.append((movie["title"], movie["rating"]["votes"]))
            
```

```python
top_actors = sorted(actors_count, key=actors_count.get, reverse=True)
for a_id in top_actors[:10]:
    print(a_id, actors_id_name_map[a_id], actors_count[a_id])
```

```python
top_genres = sorted(genres_count, key=genres_count.get, reverse=True)
for g_id in top_genres[:10]:
    print(g_id, genres_count[g_id])
```

```python
sorted(film_ratings, key=lambda v: v[1], reverse=True)[:10]
```

```python
sorted(film_votes, key=lambda v: v[1], reverse=True)[:10]
```

## Genre-Bending Hits

```python
genre_pair_rows = []

# open the ratings data
with open("../data/imdb_movies_2000to2022.prolific.json", "r") as in_file:
    
    # each line in the file is a JSON item
    for line in in_file:
        
        # read this movie
        movie = json.loads(line)

        if len(movie["rating"]) == 0:
            continue

        # Need at least double-genre movies
        if len(movie["genres"]) < 2:
            continue

        rating = movie["rating"]["avg"]
        
        sorted_genres = sorted(movie["genres"])
        for i in range(len(sorted_genres)):
            for j in range(i+1,len(sorted_genres)):
                duo_genre = "%s-%s" % (sorted_genres[i], sorted_genres[j])

                genre_pair_rows.append({
                    "genre": duo_genre,
                    "rating": rating
                })
```

```python
double_genre_df = pd.DataFrame(genre_pair_rows)
```

```python
double_genre_df.groupby("genre").mean().sort_values(by="rating", ascending=False).head(10)
```
