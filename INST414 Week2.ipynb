```python
import pandas as pd 
import json
```

```python
actors_id_name_map = {}
actors_count = {}
genres_count = {}
film_ratings = []
film_votes = []

# open the ratings data
with open("/Users/adronallen/Downloads/imdb_movies_2000to2022.prolific.json", "r") as in_file:
    
    # each line in the file is a JSON item
    for line in in_file:
        
        # read this movie
        movie = json.loads(line)
        
        # Iterate over each actor
        for actor_id,actor_name in movie["actors"]:
            actors_id_name_map[actor_id] = actor_name
            actors_count[actor_id] = actors_count.get(actor_id, 0) + 1

        # Iterate over each genre
        for genre in movie["genres"]:
            genres_count[genre] = genres_count.get(genre, 0) + 1

        if len(movie["rating"]) > 0:
            # Add the movie rating to the list
            film_ratings.append((movie["title"], movie["rating"]["avg"]))
    
            # Add the movie rating COUNT to the list
            film_votes.append((movie["title"], movie["rating"]["votes"]))
            
```

```python
# top ten most frequent actors in the dataset
top_actors = sorted(actors_count, key=actors_count.get, reverse=True)
for a_id in top_actors[:10]:
    print(a_id, actors_id_name_map[a_id], actors_count[a_id])
```

```python
# top ten most frequent genres in the dataset
top_genres = sorted(genres_count, key=genres_count.get, reverse=True)
for g_id in top_genres[:10]:
    print(g_id, genres_count[g_id])
```

```python
#top ten highest rated films in the dataset
sorted(film_ratings, key=lambda v: v[1], reverse=True)[:10]
```

```python
#top ten films with the most votes in the dataset
sorted(film_votes, key=lambda v:v[1], reverse=True)[:10]
```
